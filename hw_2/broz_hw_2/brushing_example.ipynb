**Here's a very simple example of brushing where we assume that different variables are related by a common index.**

```python
import pandas as pd
import numpy as np

from bokeh.plotting import *
from bokeh.models import ColumnDataSource

output_notebook()
```

**Make an arbitrary dataframe for testing.**

```python
x0 = np.random.rand(1000)
x1 = np.random.randn(1000)
x2 = np.sin(np.linspace(0,7,1000))
x3 = np.cos(np.linspace(0,7,1000))
x4 = np.exp(np.linspace(0,10,1000))
x5 = np.linspace(0,50,1000)

d = dict(x0=x0, x1=x1, x2=x2, x3=x3, x4=x4, x5=x5)
df = pd.DataFrame(d)
```

```python
def brushing_plot(dataframe, xylist, no_grid_cols=3):
    '''
    Takes a pandas dataframe and makes a grid of plots of the relationships
    defined in xylist as tuples. The number of columns in the grid is set by 
    key word argument: no_grid_cols.
    '''
    
    # probably there's a better way to do this
    dfdict = dataframe.to_dict()
    data = dict()
    for key in dfdict.keys():
        data[key] = dataframe[key].tolist() 
    
    source = ColumnDataSource(data=data)
    
    TOOLS = "pan,wheel_zoom,box_zoom,reset,save,box_select,lasso_select"
    
    plotdict = dict()
    for i, xy in enumerate(xylist):
        plotdict[i] = figure(tools=TOOLS, width=250, height=250, title=None)
        plotdict[i].scatter(xy[0], xy[1], source=source)
        
    gridlist = list()
    i = 0
    templist = list()
    for val in plotdict.values():
        i += 1
        templist.append(val)
        if i == 3:
            gridlist.append(templist)
            templist = list()
            i = 0
    
    p = gridplot(gridlist)

    show(p)
    
```

```python
brushing_plot(df, [("x0", "x1"), ("x1", "x2"), ("x2", "x3"), 
                   ("x3", "x4"), ("x4", "x5"), ("x2", "x4")])
```
